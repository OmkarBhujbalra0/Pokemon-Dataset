```python
names = data['Name'].to_list()
names
```

```python
r = requests.get(f'https://pokeapi.co/api/v2/pokemon-form/bulbasaur/').json()
```

```python
list_names = data['Name'].str.lower().to_list()
list_names
```

```python
form_name_available = []
form_name_notavailable = []
for name in names:
  res = requests.get(f'https://pokeapi.co/api/v2/pokemon-form/{name}/')
  if res.status_code!=200:
    print(f'{name} not available')
    form_name_notavailable.append(name)
  elif res.status_code==200:
    print(f'{name} available')
    form_name_available.append(name)
```

```python
form_name_available
```

```python
for i in range(0,len(form_name_available)):
  form_name_available[i] = form_name_available[i].lower()
len(form_name_available)
```

```python
l = []
for name in form_name_available:
  res = requests.get(f'https://pokeapi.co/api/v2/pokemon-form/{name}/').json()
  form_name = res['form_name']
  actual_name = res['pokemon']['name']
  row = {
      'actual_name':actual_name,
      'form':form_name
  }
  l.append(row)
  form_list = pd.DataFrame(l)
  form_list
```

```python
form_list.to_csv('pokeformslist.csv')
```

```python
final_data = data.merge(form_list,left_on='Name',right_on='actual_name',how='left')
final_data
```
